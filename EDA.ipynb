## Reading in the data

```python
targets = ['SOC', 'pH', 'Ca', 'P', 'Sand']
```

```python
df_train = pd.read_csv('../africa-data/training_df.csv',sep='\t')
df_train.head()
```

### Creating a dataframe WITHOUT the spectroscopy data

```python
non_spec_columns = [i for i in df_train.columns if not i.startswith('m')]
non_spec_columns = [i for i in non_spec_columns if i != 'PIDN']
non_spec_training_df = df_train[non_spec_columns]
```

## Plots of target variables

```python
for t in targets:
    fig = plt.figure(figsize=(25,5),dpi=3200)
    plot(df_train[[t]])
    legend([t])
```

## Scatter matrix of all non-spectroscopy data

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(non_spec_training_df,alpha=0.03, figsize=(30, 30), diagonal='kde')
```

## Correlation heatmap

```python
figure(figsize=(20,20),dpi=2000)
pcolor(non_spec_training_df.corr(method='pearson'),cmap='hot')
colorbar()
yticks(range(21),list(non_spec_training_df.columns.values))
xticks(range(21),list(non_spec_training_df.columns.values),rotation=50)
show()
```
